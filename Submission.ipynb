```python
drop_cols = ['sportMode', 's1', 's5', 's10', 's16', 's18', 's19']
features_col_name = raw_train.columns.drop(drop_cols).drop(["vehicleId", "days"])

raw_train.drop(drop_cols, axis=1, inplace=True)
raw_test.drop(drop_cols, axis=1, inplace=True)
```

```python
rul_test = pd.DataFrame(raw_test.groupby('vehicleId')['days'].max()).reset_index()
rul_test.columns = ['vehicleId', 'max']
rul_test.head()

## todo:::
raw_truth['rtf'] = raw_truth['RUL'] + rul_test['max']
raw_truth.head()


raw_truth.drop('RUL', axis=1, inplace=True)
raw_test = raw_test.merge(raw_truth, on=['vehicleId'], how='inner')
raw_test['ttf'] = raw_test['rtf'] - raw_test['days']
raw_test.drop('rtf', axis=1, inplace=True)
raw_test.head()


raw_train['ttf'] = raw_train.groupby(['vehicleId'])['days'].transform(max) - raw_train['days']
raw_train.head()

test_df = raw_test.copy()
train_df = raw_train.copy()
```

```python
window = 30

train_df['label'] = train_df['ttf'].apply(lambda x: 1 if x <= window else 0)
test_df['label'] = test_df['ttf'].apply(lambda x: 1 if x <= window else 0)

target_col_name = 'label'

seq_size = 60
seq_cols = features_col_name
nb_features = len(seq_cols)
batch_size = 256
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

train_df[features_col_name] = scaler.fit_transform(train_df[features_col_name])
test_df[features_col_name] = scaler.transform(test_df[features_col_name])
```

```python
def generate_ts(id_df, seq_length, seq_cols):
    df_zeros = pd.DataFrame(np.zeros((seq_length-1, id_df.shape[1])), columns=id_df.columns)
    id_df = df_zeros.append(id_df, ignore_index=True)
    data_array = id_df[seq_cols].values
    num_elements = data_array.shape[0]
    ts_array = []
    for start, stop in zip(range(0, num_elements-seq_length), range(seq_length, num_elements)):
        ts_array.append(data_array[start:stop, :])
    return np.array(ts_array)


def generate_ts_label(id_df, seq_length, seq_cols, label_col):
    df_zeros = pd.DataFrame(np.zeros((seq_length-1, id_df.shape[1])), columns=id_df.columns)
    id_df = df_zeros.append(id_df, ignore_index=True)
    data_array = id_df[seq_cols].values
    num_elements = data_array.shape[0]
    label = []
    for start, stop in zip(range(0, num_elements-seq_length), range(seq_length, num_elements)):
        label.append(id_df[label_col][stop])
    return np.array(label)
```

```python
x_train = np.concatenate(list(list(generate_ts(train_df[train_df['vehicleId'] == id],
                                               seq_size, seq_cols)) for id in train_df['vehicleId'].unique()))
y_train = np.concatenate(list(list(generate_ts_label(train_df[train_df['vehicleId'] == id],
                                                     seq_size, seq_cols, "label")) for id in train_df['vehicleId'].unique()))
x_test = np.concatenate(list(list(generate_ts(test_df[test_df['vehicleId'] == id],
                                              seq_size, seq_cols)) for id in test_df['vehicleId'].unique()))
y_test = np.concatenate(list(list(generate_ts_label(test_df[test_df['vehicleId'] == id],
                                                    seq_size, seq_cols, "label")) for id in test_df['vehicleId'].unique()))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from keras.callbacks import EarlyStopping
from keras.layers import Dense, Input, Conv1D, MaxPooling1D, LSTM, Flatten
```

```python
def make_model():
    model = Sequential()
    model.add(Conv1D(35, kernel_size=3, input_shape=(seq_size, nb_features)))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(20, kernel_size=3))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(30, activation="relu"))    
    model.add(Dense(1, activation="sigmoid"))
    return model

model = make_model()
```

```python
model.compile(loss=focal_loss, optimizer='adam', metrics=['accuracy', f1])

model.summary()

model.fit(x_train, y_train, epochs=20, batch_size=batch_size, validation_split=0.4, verbose=1,
          callbacks=[EarlyStopping(monitor='val_loss', min_delta=0, patience=30, verbose=0, mode='auto')])
```

```python
train_scores = model.evaluate(x_train, y_train, verbose=1, batch_size=batch_size)
y_prediction_train = model.predict_classes(x_train)

print('Train Accuracy: {}'.format(train_scores[1]))
print('Train F1: {}'.format(train_scores[2]))
print('Train Confusion Matrix: \n',confusion_matrix(y_train,y_prediction_train))



test_scores = model.evaluate(x_test, y_test, verbose=1, batch_size=batch_size)
y_prediction_test = model.predict_classes(x_test)

print('Test Accuracy: {}'.format(test_scores[1]))
print('Test F1: {}'.format(test_scores[2]))
print('Test Confusion Matrix: \n',confusion_matrix(y_test,y_prediction_test))
```
